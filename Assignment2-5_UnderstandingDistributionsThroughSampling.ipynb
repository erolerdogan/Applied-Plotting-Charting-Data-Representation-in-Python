```python
%matplotlib notebook
from matplotlib import pyplot as plt
from matplotlib import animation as animation
import numpy as np

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(-2.5, 1, 1000)
x2 = np.random.gamma(2, 1.5, 1000)
x3 = np.random.exponential(2, 1000)
x4 = np.random.uniform(1,7, 1000)

fig, axs = plt.subplots(2,2, figsize=(10,6),sharex=True, sharey=True)
(ax1, ax2), (ax3, ax4) = axs

values = [x1, x2, x3, x4]
axes = [ax1, ax2, ax3, ax4]
bins = np.arange(-7,8,0.5)
titles = ["Normal", "Gamma", "Exponential", "Uniform"]


def update(curr):
    
    if curr == 100:
        animat.event_source.stop()
        
    for i in range(len(axes)):
        axes[i].cla()
        axes[i].hist(values[i][:100*curr], bins=bins, density=True, alpha=.5, color="r")
        axes[i].axis([-7, 8, 0, 0.7])
        axes[i].set_title("'" + str(titles[i]) + "' the Distribution")

        fig.suptitle('Understanding Distributions Through Sampling')
        plt.subplots_adjust(hspace=0.5)        

        
animat = animation.FuncAnimation(fig, update, interval=100)
```
